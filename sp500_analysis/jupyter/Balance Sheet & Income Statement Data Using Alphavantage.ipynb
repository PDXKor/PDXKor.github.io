### Drop Columns

Alphavantage tends to put the string 'None' when no data exists, therefore it is important to inspect columns that have a high degree of nones, it may be better to drop the entire column. 

```python
missing_vals = []
for col in df.columns:
    if df[col][df[col]=='None'].shape[0] > 0:
        miss_dict = {'col_name':col, 
         'count_miss': df[col][df[col]=='None'].shape[0],  
         'pct_miss': (df[col][df[col]=='None'].shape[0] / df.shape[0])
        }
        missing_vals.append(miss_dict)

miss_df = pd.DataFrame(missing_vals)
miss_df.sort_values(by=['pct_miss'],ascending=False)
```

```python
drop_cols = miss_df['col_name'][miss_df['pct_miss']>=.3].values
df = df.drop(columns=drop_cols)
```

## Add Financial Metrics

Add additonal financial metrics, typically ratios that help us understand the proportion of certain balance sheet items relative to others. 


### Debt to Equity Ratio

The debt to equity ratio measures how much a company is funding its operations using borrowed money. It can indicate whether shareholder equity can cover all debts. 

```python
df['debtToEquity'] = df['totalLiabilities'] / df['totalShareholderEquity']
fig = px.scatter(df,x=df['totalLiabilities'],y=df['totalShareholderEquity'],hover_data=['symbol','debtToEquity','fiscalDateEnding'])
fig.show()
```

### Total Debt to Total Assets

Calculated by (Short Term Debt / Long Term Debt) / Total Assets

```python
df['totalDebtToAssets'] = (df['shortTermDebt'] + df['longTermDebt']) / df['totalAssets']
fig = px.scatter(df,x=(df['shortTermDebt'] + df['longTermDebt']) , y=df['totalAssets'],hover_data=['symbol','totalDebtToAssets','fiscalDateEnding'])
fig.show()
```

### Current Ratio

Current ratio = Current Assets / Current Liabilities. This is a measure of if a company can pay their bills in the short term.  ratio of 1.5 or greater is considered adequate.

```python
df['currentRatio'] = (df['totalCurrentAssets']) / df['totalCurrentLiabilities']
```

```python
fig = px.scatter(df,x='totalCurrentAssets',y='totalCurrentLiabilities',hover_data=['symbol','currentRatio','fiscalDateEnding'])
fig.show()
```

### Return on Equity

Return on equity is a companies net income divided by shareholders equity.

```python
df['returnOnEquity'] = (df['netIncome']) / df['totalShareholderEquity']
```

### Return on Assets

Indicates how profitable a company is in relation to its total assets. Calculated as Net Income / Total Assets

```python
df['returnOnAssets'] = (df['netIncome']) / df['totalAssets']
```

### Gross Margin

Calcualted by grossprofit / net sales

```python
df['grossMargin'] = (df['grossProfit']) / df['totalRevenue']
```

### Profit Margin

net income / net sales

```python
df['profitMargin'] = (df['netIncome']) / df['totalRevenue']
```
